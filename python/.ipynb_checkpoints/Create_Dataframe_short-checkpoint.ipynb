```python
from pj_funcs import *
import time
```

```python
reader_path = '../data/processed/column_guide_with_key.csv'
data_path = '../data/raw/export.csv'


data = pd.read_csv(data_path, encoding='Cp1252')
reader = pd.read_csv(reader_path)

# create empty dataframe
empty = pd.DataFrame(columns=reader['Processed Column'])
empty['Message'] = data['MESSAGE']
```

```python
b = HL7_PHESS_parser(reader,data)
```

```python
list(reader['Keywords'])
```

```python
#########################################################
# Find ALL
#########################################################

a = time.time()

# Loop through the messages to read and parse
for z in np.arange(0,len(data)):
    message = empty['Message'][z]
    m = hl7.parse(message)
    
    # Loop through rows in NSSP priority list
    for i in np.arange(0,len(reader)):
        row = reader.iloc[i]
        col_name = str(row['Processed Column'])
        
        # depending on variable "type" do something else
        if row['Type'] == 0:
            output = quick_grab(m,row)
            if output != output:
                output = quick_grab(m,row,SEG='1')
        
        elif row['Type'] == 1:
            output = OBX_grab1(m,row)
            if output != output:
                output = quick_grab(m,row,SEG='1')
                
        else:
            output = OBX_grab2(m,row)
            
        empty.loc[z,col_name] = output
        
        
b = time.time()

print(b-a)
```

```python
empty['Fac_name'] = data['FACILITY_NAME']
```

```python
empty.to_csv('../data/processed/2020_07_22_query.csv', index = False)
```
