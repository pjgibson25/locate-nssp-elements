```python
# Import[ant] libaries @('_')@
import hl7
import pandas as pd
import numpy as np
import re
import os
import math

###################################################

def NoError(func, *args, **kw):
    try:
        func(*args, **kw)
        return True
    except Exception:
        return False
    
def index(m,ind):
    return m[ind]


base_dir = os.getcwd()
```

```python
# Get PHESS sample
os.chdir('../data/raw')
data = pd.read_csv('PHESSsample.csv')

# Get File to read
os.chdir('../processed/')
reader = pd.read_csv('column_guide_with_key.csv')

# Change to original basedir
os.chdir(base_dir)
```

```python
def quick_grab(m,row,SEG='0'):
    
    if SEG == '0':
        end = 3
    elif SEG == '1':
        end = 1
    
    out = ''
    Segment = 'Seg'+SEG
    seg_name = row[Segment]
    
    # Make sure that the message can be indexed by segment name (not applicable to OBX)
    if NoError(index,m,seg_name):
        seg = m[seg_name][0]     
        
        # Go through each item (0_0 , 0_1 , and 0_2) to see if we find it
        for i in np.arange(0,end):
            col_title = SEG+'_'+str(i)
            num = row[col_title]
             
            # See if there is a non-null value for that row in the reader-key
            if ~np.isnan(num) == True:
                num_txt_split = str(num).split('.')
                ind1 = int(num_txt_split[0])
                ind2 = int(num_txt_split[1])-1
                
                # Can we index the segment into its field?
                if NoError(index,seg,ind1):
                    fi = seg[ind1][0]
                    
                    # Can we index the field into a component?
                    if NoError(index,fi,ind2):
                        
                        #account for odd cases (datetime column)
                        if type(fi) == str:
                            out.append(fi)
                        else:
                            out.append(str(fi[ind2]))
                            
    # make it a string array and combine                        
    out = np.array(out).astype(str)
    joined = '|'.join(out)
    without = ''.join(out)
    
    # Make sure output is not empty
    if len(without)>0:
        return joined
    else:
        return np.nan
```

```python
def OBX_grab1(m,row):
    out = []
    
    # Make sure that the message can be indexed by segment name (not applicable to OBX)
    if NoError(index,m,'OBX'):
        seg = m['OBX']
        pat = str(row['Keywords'])
        pattern = re.compile(pat,re.M|re.I)
        
        # Use regex to search for keyword matches in each OBX segment
        for i in np.arange(0,len(seg)):
            OBX_index = seg[i]
            searchme = str(OBX_index)
            match = re.findall(pattern, searchme)
            num_matches = len(match)
            
            if num_matches > 0:
                # Go through each item (0_0 , 0_1 , and 0_2) to see if we find it
                for j in np.arange(0,3):
                    col_title = '0_'+str(j)
                    num = row[col_title]
                    # See if there is a non-null value for that row in the reader-key
                    if num == num:
                        # Can we index the segment into its field?  If yes, append string to output
                        if NoError(index,OBX_index,int(num)):
                            out.append(str(OBX_index[int(num)]))

                            
    # make it a string array and combine                        
    out = np.array(out).astype(str)
    joined = '|'.join(out)
    without = ''.join(out)
    
    # Make sure output is not empty
    if len(without)>0:
        return joined
    else:
        return np.nan
```

```python
def OBX_grab2(m,row):
    out = []
    
    # Make sure that the message can be indexed by segment name (not applicable to OBX)
    if NoError(index,m,'OBX'):
        seg = m['OBX']
        pat = str(row['Keywords'])
        pattern = re.compile(pat,re.M|re.I)
        
        # Use regex to search for keyword matches in each OBX segment
        for i in np.arange(0,len(seg)):
            OBX_index = seg[i]
            searchme = str(OBX_index)
            match = re.findall(pattern, searchme)
            num_matches = len(match)
            
            # Copy search for 
            if num_matches > 0:
                match2 = match.copy()
                a = np.array([len(match2[0][q])>0 for q in np.arange(0,len(match2[0]))])
                units = np.array((match2[0]))[a][0].upper()
                out.append(units)
                
    # make it a string array and combine                        
    out = np.array(out).astype(str)
    joined = '|'.join(out)
    without = ''.join(out)
    
    # Make sure output is not empty
    if len(without)>0:
        return joined
    else:
        return np.nan
```

```python
empty = pd.DataFrame(columns=reader['Processed Column'])
empty['Message'] = data['MESSAGE']


#########################################################
# Find ALL
#########################################################

# Loop through the messages to read and parse
for z in np.arange(0,len(data)):
    message = empty['Message'][z]
    m = hl7.parse(message)
    
    # Loop through rows in NSSP priority list
    for i in np.arange(0,len(reader)):
        row = reader.iloc[i]
        col_name = str(row['Processed Column'])
        
        if row['Type'] == 0:
            output = quick_grab(m,row)
            if output != output:
                output = quick_grab(m,row,SEG='1')
        
        elif row['Type'] == 1:
            output = OBX_grab1(m,row)
            if output != output:
                output = quick_grab(m,row,SEG='1')
                
        else:
            output = OBX_grab2(m,row)
            
        empty.loc[z,col_name] = output
```
