```python
from pj_funcs import *
from itertools import compress


def percents(totals,counts):
    frac = pd.DataFrame(columns=list(counts.columns))
    for i in counts.columns:
        count_series = counts[i]
        div_series = count_series/totals
        frac[i] = div_series*100
    return frac
```
