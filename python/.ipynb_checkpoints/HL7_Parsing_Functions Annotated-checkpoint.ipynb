### Description:

Usign the column_guide_with_key.csv file, we will parse HL7 messages within a PHESS dataframe column to get specific elements.  To do this, we will use 1 main function with 3 supporting functions.  See markdowns for each function.

```python
# Import[ant] libaries @('_')@
%config Completer.use_jedi = False
import hl7
import pandas as pd
import numpy as np
import re
import os
import math
import time

###################################################

# NoError returns True if the function ran on a variable returns no error.  If an error is present, return False
def NoError(func, *args, **kw):
    try:
        func(*args, **kw)
        return True
    except Exception:
        return False 
    
# Indexing function just returns m indexed by ind.  m[ind]
def index(m,ind):
    return m[ind]


'''

The Following is frequently used:

    NoError(index, a , b)
        -Returns True if a[b] does not cause an error
        -Returns False if a[b] DOES result in an error
'''


###################################################

# Common practice to document current directory
base_dir = os.getcwd()
```

### Supporting Funciton 1 - Quick_Grab ( Message , column_guide_with_key.csv row )

For all rows in the column_guide_with_key.csv file that have $Type=0$, we will use the following function.

Quick grab addresses the most simple case:  A very clearly labeled HL7 location of the desired element.  Take the following image.  




See the following code annotation and use this 'Reader' file row / column_titles for clarification.
* 'None' and 'np.NaN' interpreted similarly


```python
def quick_grab(m,row,SEG='0'):
    
    # Check to see what segment was initiated in user-args.  Speficy number of sub_cols to loop through later
    if SEG == '0':
        end = 3
    elif SEG == '1':
        end = 1
    #######  In all examples below, we will assume the case where SEG is 0.  
    
    
    # Initiate output as empty string
    out = ''
    
    # trigger will later tell if: no component was found AND we need to add pipes for uniformity with other column vals
    trigger = 0
    
    # Specify segment with string text.  Will be later interpreted as a column in 'Reader' file. See user-args if confused.
    Segment = 'Seg'+SEG
    
    # Segment name is defined by cell value within 'Reader' file row.  Under the column Seg0 or Seg1
    seg_name = row[Segment]
    
    # See if we can index our HL7 parsed message into it's segment name (check for error)  Ex: message['PV2'] denoted m[seg_name]
    if NoError(index,m,seg_name):
        
        # If no error found, define our segment.  second index (0) inserted because of HL7 notation.
        seg = m[seg_name][0]     
        
        # We could have between 1 and 3 components composing a single NSSP priority element.  Loop through this max number (end)
        for i in np.arange(0,end):
            
            # Search the 'Reader' file row for the column labeled 0_0, 0_1, 0_2 as function passes through loop respectively/
            #      Call this number num.  Represents 3.1 in the example PV2.3.1
            col_title = SEG+'_'+str(i)
            num = row[col_title]
             
            # If the cell value in the column 0_[0,1,or 2] is non-null, investigate further.
            #       Note that many cases WILL be null if multiple components are not required.
            if ~np.isnan(num) == True:
                
                # If one NSSP Priority Element requires more than one component, this will separate them by a '|' character.
                if i > 0:
                    out += '|'
                    
                # Split the number into two groups.  Previous example (PV2.3.1) would split 3.1 into 3 and 0.  
                #      See below for why we use 0 and not 1.
                num_txt_split = str(num).split('.')
                ind1 = int(num_txt_split[0])
                
                # ****Python indexing combined with HL7 parsing output requires we subtract one from this index
                ind2 = int(num_txt_split[1])-1
                
                # Check to see if we can index our HL7 segment into a field.
                if NoError(index,seg,ind1):
                    
                    # If no error, index the segment into a field. Again, second index (0) inserted because of HL7 notation.
                    fi = seg[ind1][0]
                    
                    # Check to see if we can index our HL7 field into a component
                    if NoError(index,fi,ind2):
                        
                        # If we have made it this far, we will append something to our output.  
                        #      Signal this by adding to the trigger.
                        trigger += 1
                        
                        #  For datetime cases, the HL7 field itself is a string and indexing further will be problematic.
                        #  We therefore add the full string to our output 
                        if type(fi) == str:
                            out += fi
                            
                        #  For all other cases, we want to index the HL7 field further into a component.  
                        #  We therefore index the field into a component and add the full string to our output
                        else:
                            out += str(fi[ind2])
                            
                            
    # If the trigger is 0 and the length of the output, one of two things is true.
    # 1.  The NSSP Priority Element is one component and we could not find it.
    # 2.  The NSSP Priority Element is >1 component, but because of code structure, we could not add '|' characters
    # For case 2, this is a problem.  We address it below
    if (trigger == 0)&(len(out)==0):
        
        # Looping through the 'Reader' rows for columns 0_[0,1,2].  Find out if cell entry in that column.
        for j in np.arange(0,3):
            col_title = '0_'+str(j)
            num = row[col_title]
            
            # If (the cell entry is non-NaN) AND (j>0) we need to add a pipe to the output.
            #      This means that there are multiple requested components within the NSSP Priority element.
            #      Even though we can't find the requested components, we still divide these empty findings with '|' characters.
            if (num == num)&(j > 0):
                if j > 0:
                    out += '|'
                            
    # In most cases, return out (the output)
    if len(out)>0:
        return out
    
    # In the case where 'The NSSP Priority Element is one component and we could not find it.', return np.nan
    else:
        return np.nan
```

### Supporting Funciton 2 - OBX_grab1 ( Message , column_guide_with_key.csv row )

For all rows in the column_guide_with_key.csv file that have $Type=1$, we will use the following function.

OBX Grab addresses the case:  any element within the OBX Segment.  Take the following image.  




See the following code annotation and use this 'Reader' file row / column_titles for clarification.
* 'None' and 'np.NaN' interpreted similarly.


```python
def OBX_grab1(m,row):
    
    # Initiate output as empty string
    out = ''
    
    # trigger will later tell if: no component was found AND we need to add pipes for uniformity with other column vals
    trigger = 0
    
    # See if we can index our HL7 parsed message even has a 'OBX' segment
    if NoError(index,m,'OBX'):
        
        # If so, define segment.  Do NOT need to further index (ex: m['OBX'][0]) because OBX acts differently than other segments.
        seg = m['OBX']
        
        # Read the 'Reader' row for a column called 'Keywords' and assign it a string value to a variable, pat.
        pat = str(row['Keywords'])
        
        # Compile our Keywords to a RegEx format that ignores newline breaks (re.M) and disregards case (re.I)
        pattern = re.compile(pat,re.M|re.I)
        
        # There are typically multiple HL7 fields within an 'OBX' section.  We'll loop through the fields.
        for i in np.arange(0,len(seg)):
            
            # assign the OBX part (ex: OBX|3|) to the name OBX_index
            OBX_index = seg[i]
            
            # interpret OBX_index as a string.  Now, no longer ugly list of lists / HL7 object
            searchme = str(OBX_index)
            
            # Search our OBX field for our keywords using RegEx function.
            match = re.findall(pattern, searchme)
            
            # See how many matches we got from the function.
            num_matches = len(match)
            
            # If we have any matches, take further action.
            if num_matches > 0:
                
                # If we match, there's a chance we'll want anywhere between 1-3 components for our single NSSP Priority Element
                for j in np.arange(0,3):
                    
                    # Use our loop to identify our 'Reader' row's cell values at columns 0_[0,1,2] respectively in the loop
                    col_title = '0_'+str(j)
                    num = row[col_title]
                    
                    # Check if the cell value is non-null.
                    if (num == num):
                        
                        # If j>0, our NSSP Priority Element has multiple elements.
                        if (j>0):
                        
                            # Therefore, we'll insert pipes after second and third components.
                            out += '|'

                            #Add to trigger to indicate this action took place
                            trigger += 1
                        
                        # See if we can index our OBX field into its component (indicated by index of int(num))
                        if NoError(index,OBX_index,int(num)):
                            
                            # If we can, add string to our output!
                            out += str(OBX_index[int(num)])
    
    
    # If the trigger is 0 and the length of the output, one of two things is true.
    # 1.  The NSSP Priority Element is one component and we could not find it.
    # 2.  The NSSP Priority Element is >1 component, but because of code structure, we could not add '|' characters
    # For case 2, this is a problem.  We address it below
    if (trigger == 0)&(len(out)==0):
        
        # Looping through the 'Reader' rows for columns 0_[0,1,2].  Find out if cell entry in that column.
        for j in np.arange(0,3):
            col_title = '0_'+str(j)
            num = row[col_title]
            
            # If (the cell entry is non-NaN) AND (j>0) we need to add a pipe to the output.
            #      This means that there are multiple requested components within the NSSP Priority element.
            #      Even though we can't find the requested components, we still divide these empty findings with '|' characters.
            if (num == num) & (j > 0):
                out += '|'

    # In most cases, return out (the output)
    if len(out)>0:
        return out
    
    # In the case where 'The NSSP Priority Element is one component and we could not find it.', return np.nan
    else:
        return np.nan
```

### Supporting Funciton 3 - OBX_grab2 ( Message , column_guide_with_key.csv row )

For all rows in the column_guide_with_key.csv file that have $Type=2$, we will use the following function.

OBX Grab2 addresses the case:  searching for age_units.  Take the following image.  



See the following code annotation and use this 'Reader' file row / column_titles for clarification.
* 'None' and 'np.NaN' interpreted similarly

```python
def OBX_grab2(m,row):
    out = ''
    
    # See if we can index our HL7 parsed message even has a 'OBX' segment
    if NoError(index,m,'OBX'):
        
        # If so, define segment.  Do NOT need to further index (ex: m['OBX'][0]) because OBX acts differently than other segments.
        seg = m['OBX']
        
        # Read the 'Reader' row for a column called 'Keywords' and assign it a string value to a variable, pat.
        pat = str(row['Keywords'])
        
        # Compile our Keywords to a RegEx format that ignores newline breaks (re.M) and disregards case (re.I)
        pattern = re.compile(pat,re.M|re.I)
        
        # There are typically multiple HL7 fields within an 'OBX' section.  We'll loop through the fields.
        for i in np.arange(0,len(seg)):
            
            # assign the OBX part (ex: OBX|3|) to the name OBX_index
            OBX_index = seg[i]
            
            # interpret OBX_index as a string.  Now, no longer ugly list of lists / HL7 object
            searchme = str(OBX_index)
            
            # Search our OBX field for our keywords using RegEx function.
            match = re.findall(pattern, searchme)
            
            # See how many matches we got from the function.
            num_matches = len(match)
            
            # If we have any matches, take further action.
            if num_matches > 0:
                
                # Create copy of 'all' the matches.  I think always 1 match.
                match2 = match.copy()
                
                # Interperate matches as array and organize by longest strings.  Most empty
                a = np.array([len(match2[0][q])>0 for q in np.arange(0,len(match2[0]))])
                
                # Take the longest/uppercase match.
                units = np.array((match2[0]))[a][0].upper()
                
                # Append units to output.
                out += units
                
    
    # In most cases, return out (the output)
    if len(out)>0:
        return out
    
    # In the case where 'The NSSP Priority Element is one component and we could not find it.', return np.nan
    else:
        return np.nan
```

## Main Funciton - HL7_PHESS_parser ( column_guide_with_key.csv , PHESS_data )

We want to look through each row in column_guide_with_key.csv because each cooresponds to an element we want to get out of the message.  

Since each element either has a `Type in [1,2,3]`, and each subfunction addresses these cases, we simply need to loop through each row, see what type it has and then give some instruction on how to fill out an empty Dataframe.


```python
def HL7_PHESS_parser(reader,data, Timed=False):

    
    '''
    1. Read in Pandas Dataframe of PHESS data.
    2. Parse data-row's HL7 Message for NSSP Priority Elements
    3. Return Dataframe of Priority elements for each Message
    
    Parameters
    ----------
    reader : pandas.DataFrame, required
        interpreted from csv file in
        ../data/processed/column_guide_with_key.csv
    data : pandas.DataFrame, required
        taken from PHESS.  Needs columns below
        ['MESSAGE','FACILITY_NAME','PATIENT_VISIT_NUMBER','PATIENT_MRN']
    *Timed : bool, optional
        If True, gives completion time in seconds
    
    Returns
    -------
    DataFrame
        A pandas dataframe object is returned as a two dimensional data
        structure with labeled axes.
        
    Requirements
    ------------
    *Libraries*
    -import hl7
    -import pandas as pd
    -import numpy as np
    -import re
    -import time
    
    *Functions*
    -NoError
    -index
    -quick_grab
    -OBX_grab1
    -OBX_grab2
    
    '''
    # Keep track of when we started the function
    start_time = time.time()
    
    # Create empty dataframe to fill and pull Message column from PHESS data
    empty = pd.DataFrame(columns=reader['Processed Column'])
    empty['MESSAGE'] = data['MESSAGE']
    empty['FACILITY_NAME'] = data['FACILITY_NAME']
    empty['PATIENT_VISIT_NUMBER'] = data['PATIENT_VISIT_NUMBER']
    empty['PATIENT_MRN'] = data['PATIENT_MRN']

    ################################################################################
    # Begin to loop through messages to read and parse
    ################################################################################
    
    # Loop through all data and therefore all messages
    for z in np.arange(0,len(data)):
        
        # read in the row's message and use HL7 functions to parse it.
        message = empty['MESSAGE'][z]
        m = hl7.parse(message)

        ################################################################################
        # Begin to loop NSSP Priority Elements from column_guide_with_key.csv file
        ################################################################################
        
        # Loop through rows of the 'Reader' file
        for i in np.arange(0,len(reader)):
            
            # Each row represents a NSSP Priority element and where to find it in an HL7 message
            row = reader.iloc[i]
            
            # The NSSP_Priority element name is found here.
            col_name = str(row['Processed Column'])

            
            # Use quick_grab if Type=0, if no result see if there is another SEG,FIELD,COMP combination in the file.
            if row['Type'] == 0:
                output = quick_grab(m,row)
                
                # If output is still np.NaN, see if we can find it when SEG=1 (represents OR in element's search path)
                if output != output:
                    output = quick_grab(m,row,SEG='1')

            
            # Use OBX_grab1 if Type=1, if no result see if there is another SEG,FIELD,COMP combination in the file.
            elif row['Type'] == 1:
                output = OBX_grab1(m,row)
                
                # If output is still np.NaN, see if we can find it when SEG=1 (represents OR in element's search path)
                if output != output:
                    output = quick_grab(m,row,SEG='1')

            # Use OBX_grab2 if Type=2
            else:
                output = OBX_grab2(m,row)

            # Write result to a cell in empty dataframe
            empty.loc[z,col_name] = output
    
    #####  Once all cells are written to, divide NSSP_Priority elements with multiple subcomponents into sub-columns
    # Example:  Facility_Type will split into [Facility_Type_0,Facility_Type_1,Facility_Type_2]
    
    # Loop through columns
    for col in empty.columns:
        
        # If the column type is a string and we are NOT looking at the message column.
        if (empty[col].dtype == 'O') & (col != 'MESSAGE'):
            
            # If the column has entries that contain the character '|', delve further.
            if (sum(empty[col].str.contains('\|'))>0):
                
                # Split all column entries by their '|' character into sub columns
                newcols = empty[col].str.split('\|',expand = True)
                
                # See how many sub-columns were created and loop through them
                for j in np.arange(0,len(newcols.loc[0])):
                    
                    # Give subcolumn same name as original column with addition of '_' and # (starting at 0 counting up by 1)
                    newname = col+'_'+str(j)
                    
                    # Append this new sub-column to our dataframe
                    empty[newname] = newcols[j]

    # Now that we have already divided things on pipes, indicate empty cells and cells with only pipes as np.NaN
    empty = empty.replace(['','|','||'],np.nan)
    
    # Keep track of when we ended
    end_time = time.time()
    
    # If the user wanted to see how much time took place, print out the time difference
    if Timed == True:
        print('Time Elapsed:   '+str(round((end_time-start_time),3))+'  seconds')
              
    # Return the Dataframe
    return empty
```
